```python
df.drop_duplicates(inplace=True)
```

```python
# Y = Customer Type
# balance and imbalance dataset
df['Customer Type'].value_counts()
```

```python
# layal cust = 80
# disloyal = 20

## loyal customers  80%

## balance data operation
```

```python
loyal_cust = df[df['Customer Type']== "Loyal Customer"]
disloyal_cust = df[df['Customer Type']== "disloyal Customer"]
```

```python
loyal_customer = loyal_cust.sample(21000)
```

```python

balance_df = pd.concat([loyal_customer,disloyal_cust],axis=0)
balance_df.head()
```

```python
cat_col = balance_df.select_dtypes(include='O')
cat_col.head()
```

```python
num_col = balance_df.select_dtypes(exclude='O')
num_col.head()
```

```python
cat_col
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
encoder = OneHotEncoder(drop='if_binary')
```

```python
# print(OneHotEncoder.__doc__)
```

```python
onehot_data = encoder.fit_transform(cat_col).toarray()
```

```python
onehot_data.shape
```

```python
column_names = list(encoder.get_feature_names_out())
column_names
```

```python
one_hot = pd.DataFrame(onehot_data,columns=column_names)
one_hot.head()
```

```python
one_hot.shape , num_col.shape
```

```python
num_col.reset_index(drop=True,inplace=True)
```

```python
final_df = pd.concat([num_col,one_hot],axis=1)
final_df.head()
```
